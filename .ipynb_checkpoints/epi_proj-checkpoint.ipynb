## Constructing an Environmental Performance Index Score using PCA 
I will be using Principle Component Analysis (PCA) to create a cumulative Environmental Performance Index (`epi`) score because PCA can reduce the dimensionality of the 33 variables (`epivars`) into just one score for better analysis. It can also help us understand which features explain most of the variance in epi. 


### Exploratory Data Analysis

```python
# Preparing the data
epivars = ['epi_tbn', 'epi_tbg', 'epi_ghp', 'epi_uwd', 'epi_usd', 'epi_pmd', 
           'epi_bhv', 'epi_pbd', 'epi_par', 'epi_ozd', 'epi_msw', 'epi_had', 
           'epi_wwt', 'epi_snm', 'epi_gib', 'epi_cha', 'epi_cda', 'epi_noa', 
           'epi_bca', 'epi_sda', 'epi_nxa', 'epi_wrs', 'epi_wmg', 'epi_agr', 
           'epi_air', 'epi_ape', 'epi_bdh', 'epi_cch', 'epi_hmt', 'epi_eh', 
           'epi_h2o', 'epi_ev', 'epi_tcl']

#envind = pd.read_csv('https://www.qogdata.pol.gu.se/data/qog_ei_sept21.csv', encoding='ISO-8859-1')
envind = pd.read_csv("qog_ei_sept21.csv", encoding='ISO-8859-1')
envind = envind[envind.year == 2019]
envind = envind[['cname', 'ccodealp'] + epivars]
envind = envind.dropna().reset_index(drop = True)
envind.head()
```

```python
envind.describe()
```

```python
sns.pairplot(envind, vars=epivars)
```

```python
#dropping country names to scale variables
envind_dropnames = envind.drop(columns = ['cname', 'ccodealp'])
envind_dropnames.head()
```

```python
#fit PCA model 
model = PCA()
model.fit(envind_dropnames)
envind_pca = model.transform(envind_dropnames)

#standardizing the result
scaler = StandardScaler()
epi_vars = scaler.fit_transform(envind_pca)

#creating the epi column 
epi = epi_vars[:,0]
envind['epi'] = epi

#standardized epi column 
envind['epi']
```

### Exploring the `epi` Score
After sorting the scores, observationally, these country groupings make sense. In general, the countries with higher `epi` scores are known to be more environmentally conscious. The countries with lower `epi` scores, in general, face more internal struggles that may make it difficult to promote more environmental policies. 

```python
envind.sort_values('epi', ascending=False).head()
```

```python
envind.sort_values('epi', ascending=False).tail()
```

### Plotting `epi`
After plotting the variance of the features, seems like the first component explains much of the variance in the data. We can plot the weights of the first component to see if there are variables that are more influencial. 

```python
# Plotting Variance of Features 
features = range(model.n_components_)
plt.figure(figsize=(9, 4))
plt.bar(features, model.explained_variance_)
plt.title("Variance Explained By Feature")
plt.xticks(features)
plt.ylabel('variance')
plt.xlabel('PCA feature')
plt.show()
```

```python
# Weights of first components
weights = model.components_[0]

# Feature names
feature_names = envind.columns[2:-1]

# Plotting Weights
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(feature_names, weights)
ax.set_xlabel('Feature')
ax.set_ylabel('Weight')
ax.set_title('Weights of First Principal Component')
plt.xticks(rotation=90)
plt.show()
```

From the graph, we can see that `epi_ghp` has greater negative effect than all other features. `epi_ghp` measures greenhouse gas emissions per capita, so it makes sense that higher measures of this variables is related to lower `epi` scores. 
The two highest features are `epi_msw` (controlled solid waste) and `epi_wmg` (waste management issue) though not by a significant amount. 

These feature reveal insight on what areas can be targeted for future policies.

## Making Clusters

```python
# Scale data 
scaler = StandardScaler()
envind_scaled = scaler.fit_transform(envind_dropnames)

# Create a KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state = 1234, n_init = 'auto')

# Fit the model to the data
kmeans.fit(envind_scaled)

# Get the labels for each point
envind['epi_cl'] = kmeans.labels_
```

## Making Predictions

```python
## Prepared data for the prediction task
predictors = pd.read_csv('https://www.qogdata.pol.gu.se/data/qog_bas_ts_jan23.csv')
covars = ['ht_colonial', 'ht_region', 'fh_status', 'wbgi_pve', 'bmr_dem', 
          'wdi_popurb', 'wdi_poprul', 'wdi_pop65', 'wdi_pop1564', 'egov_egov', 'wdi_migration', 
          'wdi_area', 'wdi_agedr', 'wdi_acel', 'wdi_elerenew', 'wbgi_gee', 'wbgi_rle', 'wdi_tele', 
          'wdi_acelu', 'wdi_araland', 'wdi_wip', 'dr_ig', 'prp_prp', 'wdi_death', 'undp_hdi',
          'wdi_acelr', 'wdi_gdpind', 'wdi_birth', 'wdi_broadb', 'wdi_internet']
predictors = predictors[predictors.year == 2015]
predictors = predictors[['ccodealp'] + covars]
predictors = predictors.dropna().reset_index(drop = True)
predictors.head()
```

```python
#merging the two since they have diff amounts of rows
only_epi_df = envind[['epi', 'ccodealp']]
new_predictors = predictors.merge(only_epi_df, on='ccodealp')
new_predictors.shape
```

```python
## Lasso Regression
lasso = Lasso(alpha = 1).fit(new_predictors[covars], new_predictors['epi'])
print('The R-squared for this regression is: ' + str(lasso.score(new_predictors[covars], new_predictors['epi'])))
plt.bar(new_predictors[covars].columns, lasso.coef_)
plt.xticks(rotation=45)
plt.show()
print(lasso.intercept_)
print(np.mean(new_predictors['epi']))
```

```python
## Lasso with Cross-Validation
lasso_cv = Lasso(max_iter = 20000000, random_state = 1234)
coefs = list()
errors = list()
CVerrors = list()
alphas = np.logspace(-6, 6, 200)
X_train, X_test, y_train, y_test = train_test_split(new_predictors[covars], new_predictors['epi'], test_size = 0.2, random_state = 12345)

for a in alphas:
    lasso_cv.set_params(alpha = a).fit(X_train, y_train)
    CVerrors.append(np.mean((lasso_cv.predict(X_test) - y_test) ** 2))
    errors.append(np.mean((lasso_cv.predict(X_train) - y_train) ** 2))

print('The alpha that minimizes the Lasso Cross-Validation MSE is: ' + str(alphas[CVerrors.index(min(CVerrors))]))
    
mses = pd.DataFrame({
    'trainMSE': errors,
    'testMSE': CVerrors}, index = alphas)
```

```python
## Lasso after Cross Validation 
lasso_cv = Lasso(alpha = 0.07663410868007446).fit(new_predictors[covars], new_predictors['epi'])
print('The R-squared for this regression is: ' + str(lasso_cv.score(new_predictors[covars], new_predictors['epi'])))
plt.bar(new_predictors[covars].columns, lasso_cv.coef_)
plt.xticks(rotation=45)
plt.show()
print(lasso_cv.intercept_)
print(np.mean(new_predictors['epi']))
```

```python
#coefficient with explanatory power (not zero)
[new_predictors[covars].columns[i] for i in range(len(new_predictors[covars].columns)) if lasso_cv.coef_[i]]
```

```python
#Deep Learning

#preparing the dataset
only_epi_cl = envind[['epi_cl', 'ccodealp']]
dl_df = predictors.merge(only_epi_cl, on='ccodealp')

# train test split
X = dl_df[covars]
X = StandardScaler().set_output(transform = 'pandas').fit_transform(X)
y = dl_df['epi_cl']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12345)

# One-hot encode labels
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
```

```python
# Define the model
model = Sequential()
model.add(Dense(22, activation='sigmoid', input_shape=(30,)))  # input layer 
model.add(Dense(22, activation='relu'))
model.add(Dense(22, activation='sigmoid'))
model.add(Dense(22, activation='relu'))
model.add(Dense(22, activation='sigmoid'))
model.add(Dense(22, activation='relu'))
model.add(Dense(22, activation='sigmoid'))
model.add(Dense(3, activation='softmax')) #output layer

#Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display a summary of the model
model.summary()
```

```python
# Fitting 60 epochs
h_callback = model.fit(X_train, y_train, epochs=60, validation_data=(X_test, y_test), verbose=0)
preds = model.predict(X_test)

# Convert the output to class labels
y_pred = np.argmax(preds, axis=1)

# Confusion Matrix
print(confusion_matrix(np.argmax(y_test, axis=1), y_pred))
print(classification_report(np.argmax(y_test, axis=1), y_pred))
```

```python
macro_f1 = f1_score(np.argmax(y_test, axis=1), y_pred, average='macro')
print("Macro F1 score:", macro_f1)
```

```python
# Looking for best number of ephochs
plt.figure()
plt.plot(h_callback.history['loss'])
plt.plot(h_callback.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```
